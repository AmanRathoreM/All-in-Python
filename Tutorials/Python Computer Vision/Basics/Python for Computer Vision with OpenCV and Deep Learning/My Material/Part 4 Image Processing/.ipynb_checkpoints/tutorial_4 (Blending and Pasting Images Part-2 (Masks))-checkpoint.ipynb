```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_1=cv2.imread('dog_backpack.jpg')
img_2=cv2.imread('watermark_no_copy.png')
img_1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
img_2=cv2.cvtColor(img_2,cv2.COLOR_BGR2RGB)
```

```python
plt.title(f"This image Shape = {img_1.shape}")
plt.imshow(img_1)
```

```python
img_2 = cv2.resize(img_2, (600,600))
plt.title(f"This image Shape = {img_2.shape}")
plt.imshow(img_2)
```

```python
x_offset = img_1.shape[1] - img_2.shape[1]
y_offset = img_1.shape[0] - img_2.shape[0]
```

```python
rows, cols, channels = img_2.shape
```

```python
rows
```

```python
cols
```

```python
channels
```

```python
regin_of_intrest = img_1[y_offset:, x_offset:]
```

```python
plt.imshow(regin_of_intrest)
```

```python
img_2_gray = cv2.cvtColor(img_2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_2_gray)
```

```python
plt.imshow(img_2_gray, cmap='gray')
```

```python
img_2_gray_inverse = cv2.bitwise_not(img_2_gray)
```

```python
plt.title(f"This image Shape = {img_2_gray_inverse.shape}")
plt.imshow(img_2_gray_inverse)
```

```python
plt.title(f"This image Shape = {img_2_gray_inverse.shape}")
plt.imshow(img_2_gray_inverse, cmap='gray')
```

```python
img_2_gray_inverse_with_white_background = np.full(img_2_gray_inverse.shape,
                                                   255,
                                                   dtype=np.uint8)
img_2_gray_inverse_with_white_background
```

```python
plt.title(f"This image Shape = {img_2_gray_inverse_with_white_background.shape}")
plt.imshow(img_2_gray_inverse_with_white_background, cmap='gray')
```

```python
img_2_gray_inverse_with_all_three_channels = cv2.bitwise_or(img_2_gray_inverse_with_white_background,
                                                            img_2_gray_inverse_with_white_background,
                                                            img_2_gray_inverse)
```

```python
plt.title(f"This image Shape = {img_2_gray_inverse_with_all_three_channels.shape}")
plt.imshow(img_2_gray_inverse_with_all_three_channels, cmap='gray')
```
