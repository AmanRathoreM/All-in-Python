```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_1=cv2.imread('dog_backpack.jpg')
img_2=cv2.imread('watermark_no_copy.png')
img_1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
img_2=cv2.cvtColor(img_2,cv2.COLOR_BGR2RGB)
```

```python
plt.title(f"This image Shape = {img_1.shape}")
plt.imshow(img_1)
```

```python
plt.title(f"This image Shape = {img_2.shape}")
plt.imshow(img_2)
```

# Blending Images of Same Size

```python
img_1=cv2.resize(img_1,(1200,1200))
```

```python
img_2=cv2.resize(img_2,(1200,1200))
```

```python
plt.title(f"Now Shape is = {img_1.shape}")
plt.imshow(img_1)
```

```python
plt.title(f"Now Shape is = {img_2.shape}")
plt.imshow(img_2)
```

```python
blended_image = cv2.addWeighted(src1 = img_1, 
                          alpha = .5, 
                          src2 = img_2, 
                          beta = .5, 
                          gamma = 0)
```

```python
plt.imshow(blended_image)
```

```python
blended_image = cv2.addWeighted(src1 = img_1, 
                          alpha = 1, 
                          src2 = img_2, 
                          beta = .6, 
                          gamma = 0)
```

```python
plt.imshow(blended_image)
```

```python
blended_image = cv2.addWeighted(src1 = img_1, 
                          alpha = .8, 
                          src2 = img_2, 
                          beta = .1, 
                          gamma = 0)
```

```python
plt.imshow(blended_image)
```

# Overlaying Small Images on Top of Large Images Using NumPy Ressignment

```python
img_1=cv2.imread('dog_backpack.jpg')
img_2=cv2.imread('watermark_no_copy.png')
img_1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
img_2=cv2.cvtColor(img_2,cv2.COLOR_BGR2RGB)
```

```python
plt.title(f"This image Shape = {img_1.shape}")
plt.imshow(img_1)
```

```python
plt.title(f"This image Shape = {img_2.shape}")
plt.imshow(img_2)
```

```python
img_2_small_image = cv2.resize(img_2,(600,600)).copy()
```

```python
img_1_large_image = img_1.copy()
```

```python
plt.title(f"Now Shape is = {img_2_small_image.shape}")
plt.imshow(img_2_small_image)
```

```python
x_offset = 0
y_offset = 0
```

```python
img_2_small_image.shape
```

```python
x_end = x_offset + img_2_small_image.shape[1]
y_end = y_offset + img_2_small_image.shape[0]
```

```python
new_image = img_1_large_image.copy()
new_image[y_offset:y_end, x_offset:x_end] = img_2_small_image
```

```python
plt.imshow(new_image)
```

```python
x_offset = 200
y_offset = 350
```

```python
x_end = x_offset + img_2_small_image.shape[1]
y_end = y_offset + img_2_small_image.shape[0]
```

```python
new_image = img_1_large_image.copy()
new_image[y_offset:y_end, x_offset:x_end] = img_2_small_image
```

```python
plt.imshow(new_image)
```
