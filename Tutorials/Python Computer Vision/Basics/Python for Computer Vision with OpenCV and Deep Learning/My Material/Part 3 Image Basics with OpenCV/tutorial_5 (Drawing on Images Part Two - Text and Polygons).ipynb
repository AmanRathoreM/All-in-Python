```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def blank_image_generator(tuple_contaning_size_of_blank_image=(900,1600,3),tuple_of_figsize=(16,9)):
    blank_img=np.zeros(tuple_contaning_size_of_blank_image,dtype=int)
    plt.figure(figsize=tuple_of_figsize)
    return blank_img
```

```python
plt.imshow(cv2.putText(blank_image_generator(),
                       text="HELLO",
                       org=(100,500),
                       fontFace=cv2.FONT_HERSHEY_DUPLEX,
                       fontScale=5,
                       color=(216, 187, 237)))
```

```python
plt.imshow(cv2.putText(blank_image_generator(),
                       text="HELLO",
                       org=(500,200),
                       fontFace=cv2.FONT_HERSHEY_DUPLEX,
                       fontScale=5,
                       color=(237, 235, 187),
                       lineType=cv2.LINE_8,
                       thickness=20))
```

# Custom Polygons

```python
custom_polygon_vertices_1=np.array([ [100,300], [200,200], [400,300], [200,400] ],
                                 dtype=int)
custom_polygon_vertices_1
```

```python
custom_polygon_vertices_1.shape
```

```python
custom_polygon_vertices_reshaped_1=custom_polygon_vertices_1.reshape((4,1,2))
custom_polygon_vertices_reshaped_1
```

```python
custom_polygon_vertices_reshaped_1.shape
```

```python
plt.imshow(cv2.polylines(blank_image_generator(),
           [custom_polygon_vertices_reshaped_1],
           isClosed=True,
           color=(190, 187, 237),
           thickness=10))
```

```python
custom_polygon_vertices_2=np.array([ [200,150],[300,400],[800,300],[600,200] ],
                                 dtype=int)
custom_polygon_vertices_2
```

```python
custom_polygon_vertices_reshaped_2=custom_polygon_vertices_2.reshape((4,1,2))
custom_polygon_vertices_reshaped_2
```

```python
plt.imshow(cv2.polylines(blank_image_generator(tuple_contaning_size_of_blank_image=(500,900,3)),
           [custom_polygon_vertices_reshaped_2],
           isClosed=True,
           color=(170, 242, 219),
           thickness=10))
```
