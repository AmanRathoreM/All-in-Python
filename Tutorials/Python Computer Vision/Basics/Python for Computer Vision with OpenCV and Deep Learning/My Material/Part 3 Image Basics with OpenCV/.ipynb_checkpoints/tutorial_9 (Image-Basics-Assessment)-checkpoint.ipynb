#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('dog_backpack.jpg')
img
```

```python
plt.imshow(img)
```

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
plt.imshow(cv2.flip(img,0)) 
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
img_empty_rect=img.copy()
img_empty_rect = cv2.rectangle(img_empty_rect,pt1=(250,400), pt2=(600,700), color=(255,0,0), thickness=10)
plt.imshow(img_empty_rect)
```

```python
plt.imshow(img)
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
custom_triangle_vertices=np.array([ [500, 300], [150, 700], [650 ,700]],
                                 dtype=int)

custom_triangle_vertices_reshaped_1=custom_triangle_vertices.reshape((-1,1,2))

img_empty_triangle=img.copy()
img_empty_triangle =cv2.polylines(img_empty_triangle,
           [custom_triangle_vertices_reshaped_1],
           isClosed=True,
           color=(0, 0, 255),
           thickness=10)
plt.imshow(img_empty_triangle)
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
custom_triangle_vertices=np.array([ [500, 300], [150, 700], [650 ,700]],
                                 dtype=int)

custom_triangle_vertices_reshaped_1=custom_triangle_vertices.reshape((-1,1,2))

img_filled_triangle=img.copy()
img_filled_triangle =cv2.polylines(img_filled_triangle,
           [custom_triangle_vertices_reshaped_1],
           isClosed=True,
           color=(0, 0, 255),
           thickness=10)
plt.imshow(img_filled_triangle)
```

```python
custom_triangle_vertices=np.array([ [500, 300], [150, 700], [650 ,700]],
                                 dtype=int)

custom_triangle_vertices_reshaped_1=custom_triangle_vertices.reshape((-1,1,2))

img_filled_triangle=img.copy()


img_filled_triangle=cv2.fillPoly(img_filled_triangle, [custom_triangle_vertices_reshaped_1], color=(0, 0, 255))

plt.imshow(img_filled_triangle)
```
