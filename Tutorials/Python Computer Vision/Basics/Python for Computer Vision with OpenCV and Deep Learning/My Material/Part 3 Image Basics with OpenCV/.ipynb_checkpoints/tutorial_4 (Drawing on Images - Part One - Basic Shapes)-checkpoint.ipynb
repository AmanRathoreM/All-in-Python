```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def blank_image_generator(tuple_contaning_size_of_blank_image=(900,1600,3),tuple_of_figsize=(16,9)):
    blank_img=np.zeros(tuple_contaning_size_of_blank_image,dtype=int)
    plt.figure(figsize=tuple_of_figsize)
    return blank_img
```

```python
blank_image_generator().shape
```

```python
plt.imshow(blank_image_generator())
```

```python
plt.imshow(cv2.rectangle(blank_image_generator(),pt1=(200,200),pt2=(400,400),color=(255,255,255)))
```

```python
plt.imshow(cv2.rectangle(blank_image_generator(),pt1=(600,200),pt2=(800,400),color=(255,255,255),thickness=10))
```

```python
plt.imshow(cv2.rectangle(blank_image_generator(),pt1=(400,200),pt2=(800,400),color=(255,255,255),thickness=10))
```

```python
plt.imshow(cv2.rectangle(blank_image_generator((500,500),tuple_of_figsize=(8,6)),pt1=(200,300),pt2=(300,400),color=(255,255,255),thickness=10))
```

```python
plt.imshow(cv2.rectangle(blank_image_generator(),pt1=(400,200),pt2=(1000,100),color=(143, 255, 251),thickness=10))
```

```python
plt.imshow(cv2.circle(blank_image_generator(),center=(400,100),radius=60,color=(191, 84, 1),thickness=10))
```

```python
plt.imshow(cv2.circle(blank_image_generator(),center=(800,600),radius=160,color=(237, 145, 228),thickness=-1))
```

```python
plt.imshow(cv2.line(blank_image_generator(),pt1=(400,700),pt2=(1000,100),color=(143, 255, 251),thickness=10))
```

```python
blank_image = blank_image_generator()
cv2.line(blank_image,pt1=(200,800),pt2=(1500,100),color=(255, 87, 129),thickness=10)
cv2.line(blank_image,pt1=(200,100),pt2=(1500,800),color=(255, 87, 129),thickness=10)
cv2.line(blank_image,pt1=(850,100),pt2=(850,800),color=(255, 87, 129),thickness=10)
plt.imshow(blank_image)
```
