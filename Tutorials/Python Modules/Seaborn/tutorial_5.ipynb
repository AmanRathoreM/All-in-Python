```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('dataset_tips.csv',usecols=['total_bill','tip','size'])
data
```

```python
total_bill=data.total_bill.sort_values()
total_bill
```

```python
bins=np.linspace(1,55,12,dtype=int)
bins
```

```python
sns.distplot(total_bill)
```

```python
sns.distplot(total_bill,bins=bins)
```

```python
sns.displot(total_bill,color='deepskyblue',aspect =2.5,bins=bins,kde=True)
plt.legend()
```

```python
sns.displot(total_bill,color='deepskyblue',aspect =2.5,bins=bins,kde=True)
plt.xticks(bins)
plt.legend()
```

```python
customization=dict(color='deepskyblue',edgecolor='r',lw=20,ls='--',alpha=.91)
sns.displot(total_bill,aspect =2.5,bins=bins,kde=True,**customization)
plt.xticks(bins)
plt.legend()
```

```python
sns.displot(total_bill,aspect =2.5,bins=bins,kde=True,rug=True,rug_kws=dict(color='k'))
plt.xticks(bins)
plt.legend()
```
