```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('dataset_tips.csv',usecols=['total_bill','tip','size'])
data
```

```python
sns.displot(data['size'])
```

```python
plt.hist(x=data['size'])
plt.show()
```

```python
sns.displot(data['total_bill'])
```

```python
sns.displot(data['tip'])
```

```python
sns.displot(data['tip'],bins=2)
```

```python
sns.displot(data['tip'],bins=100)
```

```python
sns.displot(data['tip'],kind='hist')#{"hist", "kde", "ecdf"}
```

```python
sns.displot(data['tip'],kind='kde')#{"hist", "kde", "ecdf"}
```

```python
sns.displot(data['tip'],kind='ecdf')#{"hist", "kde", "ecdf"}
```

```python
sns.displot(data['tip'],kind='hist',kde=True)
```

```python
sns.displot(data['tip'],kind='hist',rug=True)
```

```python
sns.displot(data['tip'],kind='hist',rug=True,kde=True)
```

```python
sns.displot(data['tip'],kind='hist',rug=True,kde=True,color='#FF668F')
```

```python
sns.displot(data['tip'],kind='hist',rug=True,kde=True,color='#FF668F',aspect =2.5)
```

```python
sns.displot(data['tip'],kind='hist',rug=True,kde=True,color='#FF668F',aspect =2.5,label='Tips Given by Customers')
plt.legend()
```
