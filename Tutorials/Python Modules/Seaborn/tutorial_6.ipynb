```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('dataset_tips.csv',usecols=['day','total_bill','tip','size','sex'])
data
```

```python
sns.barplot(x=data.total_bill)
```

```python
sns.barplot(x=data.tip)
```

```python
sns.barplot(x=data.day, y=data.total_bill)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex',data=data)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex',data=data,capsize=.12)
```

```python
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill', hue='sex',data=data,capsize=.22,order=order)
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
           )
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.max)
plt.title('Bar Graph With Max Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=50)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=10)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(y='day', 
            x='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='total_bill', 
            y='size', 
            hue='sex',
            data=data,capsize=.22,
            estimator=np.min,
            ci=100,
            n_boot=100,
            orient='h')
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100,
            palette='winter_r')
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100,
            palette='winter_r',
            saturation=20)
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100,
            palette='winter_r',
            saturation=20,
            errcolor='#B00B69')
plt.title('Bar Graph With Min Bill Values')
```

```python
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            order=order,
            hue_order=['Female','Male'],
            estimator=np.min,
            ci=100,
            n_boot=100,
            palette='winter_r',
            saturation=20,
            errcolor='#B00B69',
            dodge=False)
plt.title('Bar Graph With Min Bill Values')
```
