```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('dataset_tips.csv',usecols=['day','total_bill','tip','size','sex'])
data
```

```python
fontdict=dict(c='#7408BB',
              fontsize=50,
              rotation=330,
              alpha=.5)
plt.figure(figsize=(16,9),
           facecolor='#FFA3BB',
           edgecolor='#82B272')
sns.barplot(x='day', 
            y='total_bill', 
            hue='sex',
            data=data,capsize=.22,
            hue_order=['Female','Male'],
            ci=100,
            n_boot=100,
            palette='winter_r',
            saturation=20,
            errcolor='#B00B69',
            dodge=False)
plt.title('More in Bar Plot',fontdict=fontdict)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```
