```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(6)
x=np.random.randint(0,999,10)
y=np.linspace(450,900,10)
x,y
```

```python
sns.lineplot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.show()
```
