```python
import pandas as pd
```

```python
sales = pd.Series([10,25,6,36,2,0,None,5], index = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun", "Mon"])
sales
```

```python
sales['Tue']=999999999
sales
```

```python
sales[1]=99
sales
```

```python
sales=sales*74.20
sales
```

```python
sales=(sales/3.33)
sales
```

```python
sales.round(2)
```

```python
sales['Mon']=9999
sales
```
